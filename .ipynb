```python
class sportmart:
    def __init__(self):
        self.inventory ={}
        self.orders ={}
    def createInventory(self, productid,product,quantity,price):
        temp ={
            "productid":productid,
            "productname":product,
            "quantity":quantity,
            "price":price
        }
        self.inventory[productid]=temp

    def createOrders(self,orderid,productid,quantity,price,total):
        temp ={
            "orderid":orderid,
            "productid":productid,
            "quantity":quantity,
            "price":price,
            "total":total
        }
        self.orders[orderid]=temp
    def printOrders(self):
        print(self.orders)
    def printInventory(self):
        print(self.inventory)
        
trinity = sportmart()
orders = open("orders.csv","r") 
o_header = orders.readline()
o_data = orders.readlines()
for data in o_data:
    tmp = data.strip().split(',')
    trinity.createOrders(tmp[0],tmp[1],tmp[2],tmp[3],tmp[4])
inventory = open("inventory.csv","r")
i_header = inventory.readline()
i_data = inventory.readlines()
for data in i_data:
    tmp = data.strip().split(',')
    trinity.createInventory(tmp[0],tmp[1],tmp[2],tmp[3])
def sportsmart():
    trinity = sportmart()

    while True:
        print("\nMenu:")
        print("1. Create New Order")
        print("2. Update Inventory")
        print("3. Display Orders")
        print("4. Display Inventory")
        print("5. Exit")

        choice = input("Enter your choice (1 2 3 4 5): ")

        if choice == "1":
            order_id = input("Enter Order ID: ")
            product_id = input("Enter Product ID: ")
            quantity = int(input("Enter Quantity: "))
            price = int(input("Enter Price: "))
            total = quantity * price
            trinity.createOrders(order_id, product_id, quantity, price, total)
            print("Order created successfully.")

        elif choice == "2":
            product_id = input("Enter Product ID: ")
            product_name = input("Enter Product Name: ")
            quantity = int(input("Enter Quantity: "))
            price = float(input("Enter Price: "))
            trinity.createInventory(product_id, product_name, quantity, price)
            print("Inventory updated successfully.")

        elif choice == "3":
            trinity.printOrders()

        elif choice == "4":
            trinity.printInventory()
        elif choice == "5":
            # Save data to CSV files before exiting
            with open("orders.csv", "a") as orders_file:
                orders_file.write("")
                for order_id, order_data in trinity.orders.items():
                    orders_file.write(f"{order_data['orderid']},{order_data['productid']},{order_data['quantity']},{order_data['price']},{order_data['total']}\n")

            with open("inventory.csv", "a") as inventory_file:
                inventory_file.write("")
                for product_id, product_data in trinity.inventory.items():
                    inventory_file.write(f"{product_data['productid']},{product_data['productname']},{product_data['quantity']},{product_data['price']}\n")

            print("Data has been exported to CSV files.")
            exit()
        else:
            print("Invalid choice. Please select a valid option.")
sportsmart()






```

```python
# Initialize empty inventory and orders dictionaries
inventory = {}
orders = {}

def add_item(item_name, quantity):
    if item_name in inventory:
        inventory[item_name] += quantity
    else:
        inventory[item_name] = quantity
    print(f"{quantity} {item_name}(s) added to inventory.")

def remove_item(item_name, quantity):
    if item_name in inventory:
        if inventory[item_name] >= quantity:
            inventory[item_name] -= quantity
            print(f"{quantity} {item_name}(s) removed from inventory.")
        else:
            print(f"Not enough {item_name} in inventory.")
    else:
        print(f"{item_name} not found in inventory.")

def place_order(item_name, quantity):
    if item_name in inventory:
        if inventory[item_name] >= quantity:
            if item_name in orders:
                orders[item_name] += quantity
            else:
                orders[item_name] = quantity
            inventory[item_name] -= quantity
            print(f"{quantity} {item_name}(s) added to orders.")
        else:
            print(f"Not enough {item_name} in inventory.")
    else:
        print(f"{item_name} not found in inventory.")

def view_inventory():
    print("Current Inventory:")
    for item, quantity in inventory.items():
        print(f"{item}: {quantity}")

def view_orders():
    print("Current Orders:")
    for item, quantity in orders.items():
        print(f"{item}: {quantity}")

# Main program loop
while True:
    print("\nInventory and Order Management Menu:")
    print("1. Add item to inventory")
    print("2. Remove item from inventory")
    print("3. Place an order")
    print("4. View current inventory")
    print("5. View current orders")
    print("6. Quit")

    choice = input("Enter your choice (1/2/3/4/5/6): ")

    if choice == '1':
        item_name = input("Enter the item name: ")
        quantity = int(input("Enter the quantity to add to inventory: "))
        add_item(item_name, quantity)
    elif choice == '2':
        item_name = input("Enter the item name: ")
        quantity = int(input("Enter the quantity to remove from inventory: "))
        remove_item(item_name, quantity)
    elif choice == '3':
        item_name = input("Enter the item name: ")
        quantity = int(input("Enter the quantity to order: "))
        place_order(item_name, quantity)
    elif choice == '4':
        view_inventory()
    elif choice == '5':
        view_orders()
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please select a valid option.")

print("Exiting the inventory and order management program.")
```

```python
class SSO:
    def __init__(self):
        self.complaints = []

    def register_complaint(self,regno,stuname,dept,title, description):
        complaint = {
            'regno':regno,
            'stuname':stuname,
            'dept':dept,
            'title': title,
            'description': description,
            'status': 'Open'
        }
        self.complaints.append(complaint)
        print(f"Complaint '{title}' registered successfully!")

    def search_complaint(self, keyword):
        matching_complaints = [complaint for complaint in self.complaints if keyword in complaint['title'] or keyword in complaint['description']]
        if matching_complaints:
            print("Matching Complaints:")
            for i, complaint in enumerate(matching_complaints):
                print(f"{i + 1}. Title: {complaint['title']}, Status: {complaint['status']}")
        else:
            print("No matching complaints found.")

    def update_status(self, title, new_status):
        for complaint in self.complaints:
            if complaint['title'] == title:
                complaint['status'] = new_status
                print(f"Status of complaint '{title}' updated to '{new_status}'")
                return
        print(f"Complaint '{title}' not found.")
    
    def display_complaints_by_regno(self, regno):
        matching_complaints = [complaint for complaint in self.complaints if complaint['regno'] == regno]
        if matching_complaints:
            print(f"Complaints for Registration Number {regno}:")
            for i, complaint in enumerate(matching_complaints):
                print(f"{i + 1}. Title: {complaint['title']}, Status: {complaint['status']}")
        else:
            print(f"No complaints found for Registration Number {regno}")

def main():
    complaint_system = SSO()
    
    while True:
        print("\nComplaint System Menu:")
        print("1. Register a Complaint")
        print("2. Search for Complaints")
        print("3. Update Complaint Status")
        print("4. Display the complaints")
        print("5. Exit")
        
        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            regno = input("enter the registraion number")
            stuname = input("enter the student name")
            dept = input("enter the department")
            title = input("Enter complaint title: ")
            description = input("Enter complaint description: ")
            complaint_system.register_complaint(regno,stuname,dept,title, description)
        elif choice == '2':
            keyword = input("Enter keyword to search for complaints: ")
            complaint_system.search_complaint(keyword)
        elif choice == '3':
            title = input("Enter complaint title to update status: ")
            new_status = input("Enter new status: ")
            complaint_system.update_status(title, new_status)
        elif choice == '4':
            regno = input("Enter registration number to display complaints: ")
            complaint_system.display_complaints_by_regno(regno)

        elif choice == '5':
            print("Exiting the complaint system")
            break
        else:
            print("Invalid choice")
if __name__ == "__main__":
    main()
```
