```python
import random

# Predefined menu with item names and prices
menu = [["cold coffee", 40], ["Tea", 10], ["coffee", 20], ["bread omelette", 60], ["chicken biryani", 120]]

# List to store orders
orders = []

def gen_order_id():
    """Generate a unique order ID based on random numbers."""
    order_id = ""
    for _ in range(5):
        order_id += str(random.randint(0, 9))
    return int(order_id)

def new_order():
    """Create a new order and add it to the orders list."""
    order_id = gen_order_id()
    order = []
    print("Menu:")
    for i, item in enumerate(menu, start=1):
        print(f"{i}. {item[0]} - ${item[1]}")
    while True:
        try:
            item_number = int(input("Enter item number (0 to finish): "))
            if item_number == 0:
                break
            elif item_number < 1 or item_number > len(menu):
                print("Invalid item number. Please choose a valid item.")
                continue
            quantity = int(input("Enter quantity: "))
            if quantity <= 0:
                print("Quantity must be greater than 0.")
                continue
            order.append({"itemnumber": item_number, "quantity": quantity})
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    
    order = {"Orderid": order_id, "items": order}
    orders.append(order)
    print(f"Order {order_id} created successfully.")

def order_bill(order):
    "Calculate the bill for a given order."
    total_bill = 0
    for item in order["items"]:
        item_number = item["itemnumber"] - 1  # Adjust for 0-based indexing
        item_price = menu[item_number][1]
        total_bill += item_price * item["quantity"]
    return total_bill

def view_orders():
    """View all orders and calculate the bill for each order."""
    if not orders:
        print("No orders available.")
        return
    for order in orders:
        print(f"Order ID: {order['Orderid']}")
        for item in order["items"]:
            item_number = item["itemnumber"] - 1  # Adjust for 0-based indexing
            item_name = menu[item_number][0]
            item_price = menu[item_number][1]
            quantity = item["quantity"]
            print(f"{item_name} x{quantity} - ${item_price * quantity}")
        total_bill = order_bill(order)
        print(f"Total Bill: ${total_bill}")
        print()

def main():
    while True:
        print("\nMenu:")
        print("1. Create new order")
        print("2. View orders")
        print("3. Exit")
        choice = input("Enter your choice: ")
        
        if choice == "1":
            new_order()
        elif choice == "2":
            view_orders()
        elif choice == "3":
            # Optionally save order details to a file here
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please choose a valid option.")

if __name__ == "__main__":
    main()
```
