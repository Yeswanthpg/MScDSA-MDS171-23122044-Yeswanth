```python
import random

# Define the predefined menu and prices
menu = {
    1: {"item": "butter chicken", "price": 30},
    2: {"item": "Tea", "price": 15},
    3: {"item": "coffee", "price": 20},
    4: {"item": "pizza", "price": 299},
    5: {"item": "biryani", "price": 199}
}

# Initialize an empty list to store orders
orders = []

# Function to create a new order
def create_order():
    order_id = generate_order_id()
    order_items = get_order_items()
    order = {"OrderID": order_id, "items": order_items}
    orders.append(order)
    print(f"Order {order_id} created successfully!")

# Function to generate a unique OrderID
def generate_order_id():
    order_id = "".join(str(random.randint(0, 9)) for _ in range(5))
    return int(order_id)

# Function to get order items and quantities
def get_order_items():
    order_items = []
    while True:
        print("Menu:")
        for item_number, item_info in menu.items():
            print(f"{item_number}. {item_info['item']} - ${item_info['price']}")
        
        choice = input("Enter the item number (0 to finish ordering): ")
        
        if choice == '0':
            break
        
        item_number = int(choice)
        
        if item_number not in menu:
            print("Invalid item number. Try again.")
            continue
        
        quantity = int(input(f"How many {menu[item_number]['item']} do you want to order? "))
        order_items.append({"itemnumber": item_number, "quantity": quantity})
    
    return order_items

# Function to view orders and calculate bills
def view_orders():
    for order in orders:
        order_id = order["OrderID"]
        order_items = order["items"]
        total_bill = 0
        
        print(f"Order {order_id}:")
        for item in order_items:
            item_number = item["itemnumber"]
            quantity = item["quantity"]
            item_name = menu[item_number]["item"]
            item_price = menu[item_number]["price"]
            item_total = item_price * quantity
            total_bill += item_total
            print(f"{item_name}: {quantity} x ${item_price} = ${item_total}")
        
        print(f"Total Bill: ${total_bill}")
        print()

# Main menu loop
while True:
    print("\nMenu:")
    print("1. Create new order")
    print("2. View Orders")
    print("3. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == '1':
        create_order()
    elif choice == '2':
        view_orders()
    elif choice == '3':
        print("Thank you for using The Enchanted Fork!")
        break
    else:
        print("Invalid choice. Try again.")
```
